```python
import pandas as pd
import os
import xml.etree.ElementTree as ET
```

```python
f = open("../Dataset/info.csv", "w")
```

```python
f.write("Namefile, Species, bbxmin, bbymin, bbxmax, bbymax\n")
for file in sorted(os.listdir('../Dataset/Information')):
    tree = ET.parse('../Dataset/Information/' + file)
    root = tree.getroot()
    stringLine = ""
    for child in root:
        if child.tag == "filename":
            stringLine = stringLine + child.text + ","
        if child.tag == "object":
            for granChild in child:
                if granChild.tag == "name":
                    stringLine = stringLine + granChild.text + ","
                if granChild.tag == "bndbox":
                    stringLine = stringLine + granChild[0].text + ","
                    stringLine = stringLine + granChild[1].text + ","
                    stringLine = stringLine + granChild[2].text + ","
                    stringLine = stringLine + granChild[3].text + "\n"
    f.write(stringLine)
    
f.close()
```
